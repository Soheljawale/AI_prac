```python
import random
import math

N = 8

# Create random board
board = [random.randint(0, 7) for _ in range(N)]

# Heuristic function
def heuristic(board):

    conflicts = 0

    for i in range(N):
        for j in range(i + 1, N):

            if board[i] == board[j]:
                conflicts += 1

            elif abs(board[i] - board[j]) == abs(i - j):
                conflicts += 1

    return conflicts

temperature = 100

# Simulated Annealing Loop
while heuristic(board) != 0 and temperature > 0:

    new_board = board.copy()

    col = random.randint(0, 7)
    row = random.randint(0, 7)

    new_board[col] = row

    current_h = heuristic(board)
    new_h = heuristic(new_board)

    delta = new_h - current_h

    # Accept better state
    if delta < 0:
        board = new_board

    else:
        probability = math.exp(-delta / temperature)

        if random.random() < probability:
            board = new_board

    temperature -= 0.1

# Output
print("Final Board:")
print(board)

print("Conflicts:", heuristic(board))
```
