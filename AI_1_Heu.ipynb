```python
import sys
```

```python
graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]
```

```python
n = len(graph) 

visited = [False] * n

current_city = 0
visited[current_city] = True

tour = [current_city]
total_cost = 0

for i in range(n-1):
    nearrst_city = -1
    min_distance = sys.maxsize

    for city in range(n):
        if not visited[city] and graph[current_city][city] < min_distance:
            min_distance = graph[current_city][city]
            nearrest_city = city
    visited[nearrst_city] = True
    tour.append(nearrst_city)
    total_cost += min_distance

    current_city = nearrst_city

# Return to starting city
    total_cost += graph[current_city][tour[0]]
    tour.append(tour[0])

    print("Tour:", tour)
    print("Minimum Cost:", total_cost)
            
            
```
