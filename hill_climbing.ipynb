```python
import random
import math
import heapq


# Function to count conflicts
def conflicts(state):
    n = len(state)
    count = 0

    for i in range(n):
        for j in range(i + 1, n):

            # Same row conflict
            if state[i] == state[j]:
                count += 1

            # Diagonal conflict
            elif abs(state[i] - state[j]) == j - i:
                count += 1

    return count


# Hill Climbing Algorithm
def hill_climbing(n, max_steps=1000):

    # Random initial state
    state = [random.randint(0, n - 1) for _ in range(n)]

    for _ in range(max_steps):

        current_conflicts = conflicts(state)

        # Goal reached
        if current_conflicts == 0:
            return state

        neighbors = []

        # Generate neighbors
        for col in range(n):
            for row in range(n):

                if row != state[col]:
                    neighbor = state.copy()
                    neighbor[col] = row
                    neighbors.append(neighbor)

        # Select best neighbor
        state = min(neighbors, key=conflicts)

        # Stop if stuck in local minimum
        if conflicts(state) >= current_conflicts:
            break

    return None


# Simulated Annealing Algorithm
def simulated_annealing(n, T=1000,
                         cooling_rate=0.99,
                         min_T=0.1):

    state = [random.randint(0, n - 1) for _ in range(n)]

    while T > min_T:

        if conflicts(state) == 0:
            return state

        # Random move
        col = random.randint(0, n - 1)
        row = random.randint(0, n - 1)

        neighbor = state.copy()
        neighbor[col] = row

        delta = conflicts(neighbor) - conflicts(state)

        # Accept better state
        # or worse state with probability
        if delta < 0 or random.random() < math.exp(-delta / T):
            state = neighbor

        # Cool temperature
        T *= cooling_rate

    return None


# A* Search Algorithm
def astar_nqueens(n):

    start = tuple(random.randint(0, n - 1)
                  for _ in range(n))

    pq = [(conflicts(start), 0, start)]

    visited = set()

    while pq:

        f, g, state = heapq.heappop(pq)

        # Goal reached
        if conflicts(state) == 0:
            return list(state)

        if state in visited:
            continue

        visited.add(state)

        # Generate neighbors
        for col in range(n):
            for row in range(n):

                if row != state[col]:

                    new_state = list(state)
                    new_state[col] = row
                    new_state = tuple(new_state)

                    if new_state not in visited:

                        h = conflicts(new_state)

                        heapq.heappush(
                            pq,
                            (g + 1 + h, g + 1, new_state)
                        )

    return None


# Main Program
n = 8

print("Hill Climbing:", hill_climbing(n))

print("Simulated Annealing:", simulated_annealing(n))

print("A* Search:", astar_nqueens(n))
```
